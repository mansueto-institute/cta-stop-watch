```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_full = pd.read_parquet("../cta-stop-etl/out/2023_cta_bus_full_day_data_v2.parquet")
```

```python
# overall how many trips are one ping over all

total_unique_trips = df_full['unique_trip_vehicle_day'].nunique()
single_row_trips = df_full.groupby('unique_trip_vehicle_day').size()
single_row_trips_count = single_row_trips[single_row_trips == 1].count()
percentage_single_row_trips = round((single_row_trips_count / total_unique_trips) * 100,3)
```

```python
print(f"There are {total_unique_trips:,} unique trips in the data")
print(f"There are {single_row_trips_count:,} unique trips in the data with one ping which is {percentage_single_row_trips:}% of the total unique trips")
```

```python
# for each pattern, how many trips with one ping and what perc of total trips

test = pd.read_parquet('../cta-stop-etl/out/all_pids_list.parquet')
pids_list = test['pid'].astype(int)

rows= []

for pid in pids_list:
    
    print(f"processing pid {pid}")

    df = pd.read_parquet(f"../cta-stop-etl/out/pids/{pid}.parquet")
    rt = df['rt'].unique()[0]
    total_unique_trips = df['unique_trip_vehicle_day'].nunique()
    single_row_trips = df.groupby('unique_trip_vehicle_day').size()
    single_row_trips_count = single_row_trips[single_row_trips == 1].count()

    percentage_single_row_trips = (single_row_trips_count / total_unique_trips) * 100

    row = {'rt': rt, 'pid': pid, 'single_row_trips': single_row_trips_count, 'percentage_single_row_trips': percentage_single_row_trips}
    rows.append(row)

one_ping_check = pd.DataFrame(rows)
```
